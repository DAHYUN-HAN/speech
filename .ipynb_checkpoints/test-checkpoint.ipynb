```python
import librosa
import matplotlib.pyplot as plt
import librosa.display
import numpy as np

import os
import sox
```

```python
speakers = ["jackson", "nicolas", "theo"]
target = 0
cnts = 5
fix_path = '../requirements/recordings/'
fmt = '.wav'


for speaker in speakers:
    file_name = fix_path + str(target) + "_" + speaker + "_" + str(cnts) + fmt
    print(file_name)
    #duration = 가져올 음성 길이 offset= n, duration = m 즉, n부터 n+m까지 음성 추출 
    #y, sr = librosa.load(file_name, duration=5)
    y, sr = librosa.load(file_name, sr=8000)
    #print('y: ', y)
    #print('sr: ', sr)

    # And compute the spectrogram magnitude and phase
    #스펙트럼은 복소수로 이루어져 있다.
    #주로 사용되는 부분은 실수 부분으로 어떤 주파수 영역의 세기(magnitude)를 의미한다.복소수 = phase
    #librosa.stft(y)를 통해 스펙트럼을 얻는다.
    S_full, phase = librosa.magphase(librosa.stft(y))#magphase()함수를 통해 두 부분을 분리할 수 있다.
    # Plot a 5-second slice of the spectrum
    print("S_full", len(S_full))
    idx = slice(*librosa.time_to_frames([0, 50], sr=sr))
    print(idx)
    plt.figure(figsize=(12, 4))
    #사람이 소리의 크기를 인지할 때, 소리의 크기가 커질수록 크기에 대한 민감도가 낮아진다.
    #소리가 작을 때는 에너지가 조금만 증가해도 소리가 커졌다고 인식하지만
    #소리가 클 때는 에너지가 훨씬 더 많이 증가해야 소리가 비슷한만큼 커졌다고 인식한다.
    librosa.display.specshow(librosa.amplitude_to_db(S_full[:, idx], ref=np.max),
                             y_axis='mel', x_axis='time', sr=sr)
    plt.colorbar()
    plt.tight_layout()
```

```python
fix_path = '../requirements/recordings/'
lists = os.listdir(fix_path)
lists.sort()
print(lists[:10])
for cnt, item in enumerate(lists):
    print(item)
    file_name = fix_path + item
    # Load an example with vocals.
    y, sr = librosa.load(file_name, duration=5)

    # And compute the spectrogram magnitude and phase
    S_full, phase = librosa.magphase(librosa.stft(y))
    # Plot a 5-second slice of the spectrum
    idx = slice(*librosa.time_to_frames([0, 50], sr=sr))
    plt.figure(figsize=(12, 4))
    librosa.display.specshow(librosa.amplitude_to_db(S_full[:, idx], ref=np.max),
                             y_axis='log', x_axis='time', sr=sr)
    plt.colorbar()
    plt.tight_layout()
    if cnt >= 10:
        break
```
